```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x_0 = 100
n = np.arange(start = 1, stop = 22)

def delta(n):
    return 2 / (2 ** n)

def method1(foo, h):
    return (foo(x_0 + h) - foo(x_0)) / h

def method2(foo, h):
    return (foo(x_0) - foo(x_0 - h)) / h

def method3(foo, h):
    return (foo(x_0 + h) - foo(x_0 - h)) / (2 * h)

def method4(foo, h):
    return ((4 / 3) * method3(foo, h)) - ((1 / 3) * (foo(x_0 + 2 * h) - foo(x_0 - 2 * h)) / (4 * h))

def method5(foo, h):
    return ((3 / 2) * method3(foo, h)) - ((3 / 5) * (foo(x_0 + 2 * h) - foo(x_0 - 2 * h)) / (4 * h)) + ((1 / 10) * (foo(x_0 + 3 * h) - foo(x_0 - 3 * h)) / (6 * h))

def foo1(x):
    return np.sin(x ** 2)

def foo2(x):
    return np.cos(np.sin(x))

def foo3(x):
    return np.exp(np.sin(np.cos(x)))

def foo4(x):
    return np.log(x + 3)

def foo5(x):
    return ((x + 3) ** 0.5)

def der_foo1(x):
    return (2 * x * np.cos(x ** 2))

def der_foo2(x):
    return (-np.sin(np.sin(x)) * np.cos(x))

def der_foo3(x):
    return (foo3(x) * np.cos(np.cos(x)) * (-np.sin(x)))

def der_foo4(x):
    return 1 / (x + 3)

def der_foo5(x):
    return (1 / 2) / (np.sqrt(x + 3))

def print_plot(foo, der_foo):
    y1 = np.abs(der_foo(x_0) - method1(foo, delta(n)))
    y2 = np.abs(der_foo(x_0) - method2(foo, delta(n)))
    y3 = np.abs(der_foo(x_0) - method3(foo, delta(n)))
    y4 = np.abs(der_foo(x_0) - method4(foo, delta(n)))
    y5 = np.abs(der_foo(x_0) - method5(foo, delta(n)))

    plt.figure(figsize=[12, 5], dpi = 100)
    plt.plot(delta(n), y1, label='method 1')
    plt.plot(delta(n), y2, label='method 2')
    plt.plot(delta(n), y3, label='method 3')
    plt.plot(delta(n), y4, label='method 4')
    plt.plot(delta(n), y5, label='method 5')    
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel("$n$")
    plt.ylabel("$log y$")
    plt.legend(loc='best', bbox_to_anchor=(1, 1))
    plt.grid()
    plt.show() 
```

```python
print_plot(foo1, der_foo1)
```

```python
print_plot(foo2, der_foo2)
```

```python
print_plot(foo3, der_foo3)
```

```python
print_plot(foo4, der_foo4)
```

```python
print_plot(foo5, der_foo5)
```
