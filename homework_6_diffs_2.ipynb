```python
import numpy as np
import matplotlib.pyplot as plt
```

Уравнение Ван-дер-Поля:

$ y_1' = y_2 $

$ y_2' = -a(y_2(y_1^2 - 1) + y_1) $

$ y_{01} = 2, \,\,\, y_{20} = 0 $

$ a = 10^3 \,\,\, или \,\,\, a = 1 $

#### Решение будет получено неявным методом Рунге-Кутты (4 порядок) ####

```python
iter_num = 10
initial = np.array([2.0, 0.0])

def func(arr, coeff):
    return np.array([arr[1], -coeff * (arr[1] * (arr[0] * arr[0] - 1) + arr[0])])

def RungeKutta(func, init_array, coeff, h, iterations):
    y = [init_array]

    b = np.array([1/2, 1/2])
    a = np.array([[1/4, 1/4 - np.sqrt(3)/6], [1/4 + np.sqrt(3)/6, 1/4]])

    for i in range(iterations):
        k = np.zeros(shape=(len(init_array), len(init_array)))
        for j in range(iter_num):
            k[0] = func(y[-1] + h*(a[0][0]*k[0] + a[0][1]*k[1]), coeff)
            k[1] = func(y[-1] + h*(a[1][0]*k[0] + a[1][1]*k[1]), coeff)
        y.append(y[-1] + h * (b[0] * k[0] + b[1] * k[1]))
    
    return y
```

```python
result = RungeKutta(func, initial, 1e3, 0.0001, 100000)
x = []
y = []
for i in result:
    x.append(i[0])
    y.append(i[1])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
result = RungeKutta(func, initial, 1, 0.0001, 100000)
x = []
y = []
for i in result:
    x.append(i[0])
    y.append(i[1])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
